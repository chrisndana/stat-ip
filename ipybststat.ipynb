```python
# encoding data using OrdinalEncoder
from numpy import asarray
from sklearn.preprocessing import OrdinalEncoder
encoder=OrdinalEncoder()
df6=encoder.fit_transform(df5)
df6
```

```python
# changing the encoded data to dataframe
df7=pd.DataFrame(df6)
df7.head()
```

```python
# renaming the columns in the new dataframe accordingly
df8=df7.rename(columns={0:'Has_a_Bank_account',1:'Type_of_Location',2:'Cell_Phone_Access',3:'household_size',4:'Respondent_Age',5:'gender_of_respondent',6:'The relationship_with_head',7:'marital_status',8:'Level_of_Educuation',9:'Type_of_Job'})
df8.head()
```

```python
# principal component analysis
# defining x and y and standardizing x
from sklearn.preprocessing import StandardScaler
x=df8.iloc[:,1:10].values
y=df8.loc[:,['Has_a_Bank_account']].values
x_std=StandardScaler().fit_transform(x)
```

```python
# spliting data into training data  and test data
from sklearn.model_selection import train_test_split
x_std_train,x_std_test,y_train,y_test=train_test_split(x_std,y,test_size=0.2,random_state=0)
```

```python
from sklearn.decomposition import PCA
pca=PCA()
x_std_train=pca.fit_transform(x_std_train)
x_std_test=pca.transform(x_std_test)

```

```python
# getting the explained variance
explained_var=pca.explained_variance_ratio_
explained_var
```

```python
# determining the number of peincipal components we will use in our analysis
pca=PCA().fit(x_std)
plt.plot(np.cumsum(pca.explained_variance_ratio_))
plt.xlim(0,9,1)
plt.xlabel('number of components')
plt.ylabel('cumulative explained variance')
```

```python
from sklearn.decomposition import PCA
pca=PCA(n_components=6)
x_std_train=pca.fit_transform(x_std_train)
x_std_test=pca.transform(x_std_test)
```

```python
from sklearn.ensemble import RandomForestClassifier
classifier=RandomForestClassifier(max_depth=2,random_state=0)
classifier.fit(x_std_train,y_train)
y_prediction=classifier.predict(x_std_test)
```

```python
explained_var=pca.explained_variance_ratio_
explained_var
```

```python
# getting the accuracy level of the model
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score
confusion_mat=confusion_matrix(y_test,y_prediction)
print(confusion_mat)
print('acuracy',accuracy_score(y_test,y_prediction))
```
